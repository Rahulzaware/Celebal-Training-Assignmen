```python
import pandas as pd
import numpy as np
import string
import os
from datetime import date
```

```python
today = date.today()
end_date = today
```

```python
search_term_list = ['aap','bsp','bjp','sp','congress']
from_date = '2021-01-01'
max_results = 1000 
```

```python
import time
start = time.time()
# your code
def extracted_tweets(search_term):
  '''Fuction to extract text f from twitter and create text files..'''
  for i in range(len(search_term)):
    os.system("snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-search '{search_term[i]} until:{end_date}' > {search_term[i]}.txt")
# end
print(f'Time: {time.time() - start}')
extracted_tweets(search_term_list)
```

```python
start = time.time()
# your code
text_files=['aap.txt','bjp.txt','bsp.txt','congress.txt','sp.txt']
def merge_text_files(text_files):
  '''Function to merge five text files into single merge file..'''
  with open('merge_file.txt', 'w') as outfile:
    for i in text_files:
      with open(i) as infile:
        for line in infile:
          outfile.write(line)
# end
print(f'Time: {time.time() - start}')
merge_text_files(text_files)
```

```python
#convert text file into csv file
df = pd.read_csv("/content/merge_file.txt",header = None,names=['text'])
df.to_csv('text_file.csv',index = None)
```

# cleaning of data

To remove usernames:

```python
#Remove the username from each row of data with different conditions...
df['text'] = df['text'].str.replace(r'\s*@\w+', '', regex=True)
df['text'] = df['text'].str.replace(r'\s*\B@\w+', '', regex=True)
df['text'] = df['text'].str.replace(r'\s*@\S+', '', regex=True)
df['text'] = df['text'].str.replace(r'\s*@\S+\b', '', regex=True)

df
```

Function to remove punctuations:

```python
# define punctuation
start = time.time()
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
# function to remove punctuation 
def remove_punctuations(row):
    '''Function to remove different punctuations from each row of data..'''
    no_punct = ""
    for i in row:
        if i not in punctuations:
            no_punct = no_punct + i
    return no_punct
print(f'Time: {time.time() - start}')
```

```python
df['text'] = df['text'].apply(remove_punctuations)
df
```

Function to remove emojis:

```python
start = time.time()
def remove_emoji(row):
    '''Function to remove emojis from each row of dataset..'''
    a = ''
    for i in row:
        b=i.encode('ascii', 'ignore').decode('ascii')
        a+=b
    return a
print(f'Time: {time.time() - start}')
```

```python
df['text'] = df['text'].apply(remove_emoji)
df
```

# Labelling

To extract specif values from text column for labelling...

```python
#create new column label and store the extracted elements i.e elements of list
df['label'] = df.text.str.lower().str.extract('(aap|bsp|bjp|sp|congress)')
```

```python
df.head(10)
```

Replace label with 0 and 1:

```python
#Replace the extracted elements with 0 and 1
df['label'] = df['label'].replace(to_replace=['aap', 'bsp','bjp','sp','congress'], value='1')
df['label'].fillna(value = 0,inplace = True)
```
